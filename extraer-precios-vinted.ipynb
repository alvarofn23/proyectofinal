```python
import pandas as pd
import re

import pandas as pd
import re
df= pd.read_csv('/kaggle/input/vinteddd/Vinted Scraper b.csv')



# Función para extraer el precio después de "tamaño"
def extract_pago_after_tamano(hover_title):
    # Buscar la palabra "tamaño", seguida de una coma y luego el primer precio
    match = re.search(r'tamaño:\s*[^,]+,\s*(\d+[,\.]?\d*)', str(hover_title))
    if match:
        # Extraer el primer precio encontrado
        precio_pago = float(match.group(1).replace(',', '.'))  # Convertir a float y reemplazar coma por punto
        return precio_pago
    return None  # Si no se encuentra el patrón, retornamos None

# Aplicar la función a la columna hover_title para crear la columna 'Pago'
df['Pago'] = df['hover_title'].apply(extract_pago_after_tamano)

# Verificar las primeras filas para asegurarnos de que 'Pago' se haya añadido correctamente
print(df[['hover_title', 'Pago']].head())

# Guardar el dataset modificado (si lo deseas, puedes guardarlo localmente o en otro lugar)
df.to_csv('/kaggle/working/Vinted_Scraper_modified.csv', index=False)

# Mostrar el nombre del archivo guardado
print("Archivo guardado como: Vinted_Scraper_modified.csv")

# Ver las primeras filas del dataset modificado
df.head()


```

```python
import pandas as pd


# Lista de marcas válidas proporcionadas
valid_brands = ['balenciaga', 'gucci', 'jean paul gaultier', 'dolce gabbana', 'number nine', 
                'dries van noten', 'acne', 'givenchy', 'prada', 'alyx', 'marc jacobs', 
                'valentino', 'mcqueen', 'yohji', 'dior', 'mihara', 'louboutin', 'miu miu', 
                'saint laurent', 'supreme', 'dsquared', 'rick owens', 'mugler', 'issey miyake', 
                'vivienne westwood', 'margiela']

# Cargar el dataset de Vinted
vinted_data = pd.read_csv("/kaggle/working/Vinted_Scraper_modified.csv")

# Filtrar las columnas necesarias
vinted_filtered = vinted_data[['hover_title', 'URL', 'Marca', 'Tamaño', 'Estado', 'Color', 'Pago', 'Descripción', 'Vendedor', 'Reseña']].copy()

# Diccionario de traducción de tipos de prendas en francés, italiano y español a inglés
translation_dict = {
    'camicia': 't-shirt',       # Italiano
    'giubbotto': 'jacket',      # Italiano
    'pullover': 'sweater',      # Italiano
    'occhiali da sole': 'sunglasses', # Italiano
    'camisa': 't-shirt',        # Español
    'chaqueta': 'jacket',       # Español
    'pantalones': 'jeans',      # Español
    'cartera': 'purse',         # Español
    't-shirt': 't-shirt',       # Inglés ya está
    'tshirt': 't-shirt',        # Variación "tshirt"
    't shirt': 't-shirt',       # Variación "t shirt"
    'jacket': 'jacket',         # Inglés ya está
    'jeans': 'jeans',           # Inglés ya está
    'purse': 'purse',           # Inglés ya está
    'knit': 'knit',             # Inglés ya está
    'sweater': 'sweater',       # Inglés ya está
    'shirt': 't-shirt',         # Añadido: "shirt" también es una variación de t-shirt
    'giacca': 'jacket',         # Italiano para chaqueta
    'lunettes de soleil': 'sunglasses',  # Francés
    'jean': 'jeans',            # Francés
    'chemise': 't-shirt',       # Francés para camisa
    'manteau': 'jacket',        # Francés para abrigo/chaqueta
    'sac': 'purse',             # Francés para bolso
    'pull': 'sweater',          # Francés para suéter
}

# Función para traducir y extraer el tipo de prenda
def translate_product_type(hover_title):
    # Convertir el título a minúsculas para evitar problemas de mayúsculas
    hover_title_lower = hover_title.lower()
    
    # Buscar el tipo de prenda en el diccionario
    for key, value in translation_dict.items():
        if key.lower() in hover_title_lower:
            return value  # Retorna la traducción al inglés del tipo de prenda
    return None  # Si no se encuentra un tipo de prenda válido

# Función para traducir la condición del producto
def translate_condition(estado):
    condition_dict = {
        'Nuevo con etiquetas': 'Never worn',
        'Muy bueno': 'Very good condition',
        'Bueno': 'Good condition',
        'Satisfactorio': 'Fair condition'
    }
    return condition_dict.get(estado, None)  # Si no coincide, devuelve None

# Aplicar la traducción para extraer el tipo de prenda
vinted_filtered.loc[:, 'product_type'] = vinted_filtered['hover_title'].apply(translate_product_type)

# Traducir la condición del producto
vinted_filtered.loc[:, 'product_condition'] = vinted_filtered['Estado'].apply(translate_condition)

# Eliminar los productos que no tengan un tipo de prenda válido o condición de producto
# Solo eliminaremos aquellos que no tengan un tipo o condición válida
vinted_filtered = vinted_filtered.dropna(subset=['product_type', 'product_condition'], how='any')

# Cambiar el nombre de la columna 'Pago' a 'price_usd' (columna de precio para Vestiaire)
vinted_filtered.rename(columns={'Pago': 'price_usd'}, inplace=True)

# Crear las nuevas columnas para el formato de Vestiaire usando .loc[] para evitar el warning
vinted_filtered.loc[:, 'product_name'] = vinted_filtered['hover_title']
vinted_filtered.loc[:, 'product_description'] = vinted_filtered['Descripción']
vinted_filtered.loc[:, 'product_keywords'] = vinted_filtered['Marca'] + " " + vinted_filtered['product_type']
vinted_filtered.loc[:, 'seller_user_name'] = vinted_filtered['Vendedor']
vinted_filtered.loc[:, 'reseña'] = vinted_filtered['Reseña']

# Agregar las columnas faltantes de Vestiaire
vinted_filtered.loc[:, 'reserved'] = False
vinted_filtered.loc[:, 'available'] = True
vinted_filtered.loc[:, 'in_stock'] = True
vinted_filtered.loc[:, 'should_be_gone'] = False
vinted_filtered.loc[:, 'brand_id'] = 'NaN'  # No disponible en el dataset de Vinted, lo podemos dejar como NaN
vinted_filtered.loc[:, 'brand_url'] = 'NaN'  # Lo mismo con la URL de la marca

# Reorganizar las columnas en el formato de Vestiaire
vinted_vestiaire = vinted_filtered[['product_name', 'product_type', 'product_description', 'product_keywords', 'product_condition', 
                                    'seller_user_name', 'reseña', 'reserved', 'available', 'in_stock', 'should_be_gone', 
                                    'brand_id', 'brand_url', 'price_usd', 'URL']]

# Exportar el dataset adaptado a un archivo CSV en Kaggle
vinted_vestiaire.to_csv('/kaggle/working/Vinted_Scraper_adapted.csv', index=False)

# Mostrar las primeras filas del dataframe resultante
vinted_vestiaire.head()

```
